```python
import cv2
import numpy as np

img = cv2.imread("./assets/digits_dataset/0_1.png")
img.shape, img
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img.shape, img
```

```python
train = np.array([])

for i in range(6):
    for j in range(1, 251):
        img = cv2.imread(f"./assets/digits_dataset/{i}_{j}.png")
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        train = np.append(train, gray)
train = train.reshape(-1, 400).astype(np.float32)

test = np.array([])

for i in range(6):
    for j in range(251, 501):
        img = cv2.imread(f"./assets/digits_dataset/{i}_{j}.png")
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        test = np.append(test, gray)
test = test.reshape(-1, 400).astype(np.float32)

k = np.arange(6)
train_labels = np.repeat(k, 250)[:, np.newaxis]
test_labels = train_labels.copy()

knn = cv2.ml.KNearest_create()
knn.train(train, cv2.ml.ROW_SAMPLE, train_labels)
ret, result, neibours, dist = knn.findNearest(test, k=5)

result.shape, result
```

```python
macthes = result == test_labels
correct = np.count_nonzero(macthes)
accuracy = correct * 100 / result.size

print(f"Accuracy: {accuracy}")
```

```python
mydigit = cv2.imread("./assets/unknown_x.png")
mydigit_gray = cv2.cvtColor(mydigit, cv2.COLOR_BGR2GRAY)
mydigit_test = mydigit_gray.reshape(-1, 400).astype(np.float32)
ret, result, neibours, dist = knn.findNearest(mydigit_test, k=5)
print(ret)

font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(mydigit, str(int(ret)), (0, 7), font, 0.3, (255, 0, 0), 1, cv2.LINE_AA)

cv2.imshow("Detected Digit", mydigit)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## การจัดกลุ่มแบบเคมีน

```python
from sklearn.metrics import accuracy_score, classification_report
import numpy as np
import cv2

img = cv2.imread(".assets/house.jpg")

Z = img.reshape((-1, 3))
Z = np.float32(Z)

criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
K = 8
ret, label, center = cv2.kmeans(Z, K, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS)

center = np.uint8(center)
res = center(label.flatten())
res2 = res.reshape((img.shape))

cv2.imshow("Res2", res2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
