```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%config InLineBackend.figure_format = "retina"
```

```python
bp = pd.read_csv('./assets/Blood_Pressure.csv')
bp.corr()
```

```python
bp.plot(x="mother blood pressure", y="baby blood pressure", style="o")
plt.xlabel("mother blood pressure")
plt.ylabel("baby blood pressure")
```

```python
from sklearn.linear_model import LinearRegression

x = bp[["mother blood pressure"]]
y = bp["baby blood pressure"]
lrm = LinearRegression()
lrm.fit(x, y)

lrm.intercept_
lrm.coef_

predictions = lrm.predict(x)
predictions
```

plt.scatter(x, y, color="black")
plt.plot(x, predictions, color="blue", linewidth=3)
plt.title("mother blood pressure vs baby blood pressure")
plt.xlabel("mother blood pressure")
plt.xlabel("baby blood pressure")


```python
lrm.predict([[130]])
```

```python
lrm.predict([[160]])
```

```python
from sklearn import metrics

print("MAE: ", metrics.mean_absolute_error(y, predictions))
print("MSE: ", metrics.mean_squared_error(y, predictions))
print("RMSE: ", np.sqrt(metrics.mean_squared_error(y, predictions)))
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

ar = pd.read_csv('./assets/Area_Rental.csv')

x = ar[["Area"]]
y = ar["Rental"]
lrm = LinearRegression()
lrm.fit(x, y)

lrm.intercept_
lrm.coef_

predictions = lrm.predict(x)

plt.scatter(x, y, color="black")
plt.plot(x, predictions, color="blue", linewidth=3)
plt.title("Area vs Rental")
plt.xlabel("Area")
plt.xlabel("Rental")
```
