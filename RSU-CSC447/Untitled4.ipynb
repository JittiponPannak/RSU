```python
import cv2
import numpy as np

img = cv2.imread('./assets/blox.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = np.float32(gray)

dst = cv2.cornerHarris(gray, 2, 3, 0.04)
img[dst > 0.01 * dst.max()] = [ 0, 0, 255 ]

cv2.imshow("Harris", img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

img = cv2.imread('./assets/coca-cola.jpg')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
template = cv2.imread('./assets/cola.jpeg', 0)
w, h = template.shape[::-1]

res = cv2.matchTemplate(img_gray, template, cv2.TM_CCOEFF_NORMED)
threshold = 0.8
loc = np.where(res >= threshold)

for pt in zip(*loc[::-1]):
    cv2.rectangle(img, pt, (pt[0] + w, pt[1] + h), (0, 255, 0), 2)

cv2.imshow("Result", img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

img = cv2.imread('./assets/Mario.png')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
template = cv2.imread('./assets/Mario.png', 0)
template = template[195:220, 140:165]

w, h = template.shape[::-1]

res = cv2.matchTemplate(img_gray, template, cv2.TM_CCOEFF_NORMED)
threshold = 0.933
loc = np.where(res >= threshold)

for pt in zip(*loc[::-1]):
    cv2.rectangle(img, pt, (pt[0] + w, pt[1] + h), (0, 0, 255), 1)

cv2.imshow("Result", img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

img = cv2.imread('./assets/house.jpg')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


sift = cv2.SIFT_create()
kp = sift.detect(img_gray, None)
img = cv2.drawKeypoints(img_gray, kp, img, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

cv2.imshow("Result", img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
img1c = cv2.imread("./assets/cola.jpeg")
img1 = cv2.cvtColor(img1c,cv2.COLOR_BGR2GRAY)
img2c = cv2.imread("./assets/coca-cola.jpg")
img2 = cv2.cvtColor(img2c,cv2.COLOR_BGR2GRAY)

sift = cv2.SIFT_create()

kp1, des1 = sift.detectAndCompute(img1, None)
kp2, des2 = sift.detectAndCompute(img2, None)
match = cv2.BFMatcher()
matches = match.knnMatch(des1,des2,k=2)
good = []
for m,n in matches:
    if m.distance < 0.5*n.distance:
        good.append(m)

matching_result = cv2.drawMatches(img1c,kp1,img2c,kp2,good,None)
cv2.imshow("Matching result", matching_result)
    
if len(good)>5:
    query_pts = np.float32([kp1[m.queryIdx].pt for m in good]).reshape(-1,1,2)
    train_pts = np.float32([kp2[m.trainIdx].pt for m in good]).reshape(-1,1,2)
    matrix, mask = cv2.findHomography(query_pts, train_pts, cv2.RANSAC, 5.0)
    matches_mask = mask.ravel().tolist()
    h, w = img1.shape
    pts = np.float32([[0,0],[0,h],[w,h],[w,0]]).reshape(-1,1,2)
    dst = cv2.perspectiveTransform(pts, matrix)
    homography = cv2.polylines(img2c,[np.int32(dst)],True, (255,0,0),2)
    cv2.imshow("Homography", homography)
cv2.waitKey(0)  
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
import cv2
import numpy as np

face = cv2.CascadeClassifier('./assets/haarcascade_frontalface_default.xml')
img = cv2.imread('./assets/lena.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = face.detectMultiScale(gray, 1.3, 5)

for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)
    roi_gray = gray[y:y + h, x:x + w]
    roi_color = img[y:y + h, x:x + w]

cv2.imshow('img', img)
cv2.waitKey(0)  
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
import cv2
import numpy as np

face = cv2.CascadeClassifier('./assets/haarcascade_frontalface_default.xml')
eye = cv2.CascadeClassifier('./assets/haarcascade_eye.xml')

img = cv2.imread('./assets/lena.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faces = face.detectMultiScale(gray, 1.3, 5)

for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 2)
    roi_gray = gray[y:y + h, x:x + w]
    roi_color = img[y:y + h, x:x + w]
    
    eyes = eye.detectMultiScale(roi_gray)

    for (ex, ey, ew, eh) in eyes:
        cv2.rectangle(roi_color, (ex, ey), (ex + ew, ey + eh), (0, 255, 0), 2)

cv2.imshow('img', img)
cv2.waitKey(0)  
cv2.destroyAllWindows()
cv2.waitKey(1)
```
