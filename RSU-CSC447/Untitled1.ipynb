# Assignment 2

```python
import cv2, os

dir = "./assets/photostock/flips/"

for root, dirs, files in os.walk(dir):
    for file in files:
        print(file)
        img = cv2.imread(dir + file)
        imgflip = cv2.flip(img, 00)
        imgflip = cv2.flip(img, -1)
        cv2.imwrite("./assignment/flip/" + file.replace(".", "_2."), imgflip)
```
