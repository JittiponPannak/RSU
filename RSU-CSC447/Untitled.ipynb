```python
import cv2

img = cv2.imread('./assets/lena.jpg')

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2.imshow('image', gray)
cv2.waitKey(0)
cv2.destryAllWindows()
```

```python
import cv2

img = cv2.imread('./assets/lena.jpg', 0)

cv2.imshow('image', img)
cv2.imwrite('./assets/lena-gray.jpg', img)
cv2.waitKey(0)
cv2.destryAllWindows()
```
