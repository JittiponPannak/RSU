## 1.2 เขียนโปรแกรมคำนวณ ดอกเบี้ยรายเดือนของเงินกู้

โดยนักศึกษาต้องกำหนด “อย่างน้อย 4 เงื่อนไข” สำหรับอัตราดอกเบี้ยที่แตกต่างกันตามยอดเงินกู้ เช่น
* ยอดเงินกู้ ≤ 10,000 บาท → ดอกเบี้ย 5%
* 10,001–20,000 บาท → ดอกเบี้ย 10%
* etc
ข้อกำหนด:
* โปรแกรมต้องรับ “จำนวนเงินกู้” (ชนิดข้อมูลเป็นจำนวนเต็ม)
* ใช้ ลูป (loop) ในการคำนวณดอกเบี้ย “หน่วยละ 1 บาท” ตามเงื่อนไขที่กำหนด
* แสดง ดอกเบี้ยรวมทั้งหมด ที่ต้องจ่าย


#### ข้อกำหนด
- โปรแกรมต้องรับ “จำนวนเงินกู้” (ชนิดข้อมูลเป็นจำนวนเต็ม)
- หากยอดเงินกูน้อยกว่า 200,000 ดอกเบี้ย 5%
- หากยอดเงินกูสูงกว่า 200,000 ดอกเบี้ย 15%
- ดอกเบี้ยเพิ่ม 5% ของจำนวนเงินกู้ ทุกๆ 100,000 บาท หลังจาก และ ตั้งแต่การยืม 500,000 บาทขึ้นไป

```python
    input_loan = int(input("Input your loan: "))
    interest = 0.0
    interest_gap = 0.0
    interest_rate = 5.0 if input_loan < 200_000 else 15.0

    if input_loan >= 500_000:
        gap = int((input_loan - 400_000) / 100_000)
    
        for i in range(gap):
            interest_gap += input_loan / 5.0

    interest += input_loan / interest_rate
    interest += interest_gap

    print("Interest to Pay: ", interest)
    print("Interest Added: ", interest_gap)
```
