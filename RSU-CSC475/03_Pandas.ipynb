# DataFrame

```python
list = [
    [1,	'Garnt', 500, 12],
    [2, 'Beer', 400, 17],
    [3, 'Min', 350, 18]]
df = pd.DataFrame(list, columns=[ 'Key', 'Name', 'Wage', 'Hour' ])
df
```

```python
df['Salary'] = df['Wage'] * df['Hour']
df
```

```python
df[(df['Wage'] >= 400) & (df['Hour'] < 15)]
```

```python
df.drop('Salary', axis=1, inplace=True)
df
```

# BMI

```python
list = [
        [ 9435, "Somsri", 95, 51, 159 ]
    ,   [ 9435, "SomSak", 75, 71, 177 ]
    ,   [ 9435, "Somboon", 90, 90, 181 ]
    ,   [ 9435, "SomJai", 42, 42, 161 ]
    ]
cols = [ 'HOSPCODE', 'NAME', 'AGE', 'WEIGHT', 'HEIGHT' ]

df = pd.DataFrame(list, columns=cols)
df['BMI'] = df['WEIGHT'] / ((df['HEIGHT'] / 100) ** 2)

df[df['BMI'] > 22]
```
