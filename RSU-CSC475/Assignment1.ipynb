```python
import pandas as pd
```

```python
data = [
    [ 'Fallout 1', 1997, 89, 8.7, 8.0 ],
    [ 'Fallout 2', 1998, 86, 9.1, 7.0 ],
    [ 'Fallout 3', 2008, 93, 8.2, 6.0 ],
    [ 'Fallout NV', 2010, 84, 8.6, 9.0 ],
    [ 'Fallout 4', 2015, 87, 7.0, 6.0 ]]
cols = [ 'Name', 'Release Date', 'Meta Score', 'User Score', 'My Score' ]

df = pd.DataFrame(data, columns=cols)

# Calculate Average Score as Overall Score
df['Overall Score'] = (((df['Meta Score'] / 10) + df['User Score'] + df['My Score']) / 30.0) * 100

# Print

print("\n===== My Score : More Than : 6.0 ======\n")
print(df[df['My Score'] > 6.0])

print("\n===== Overall Score : Less Than : 80.0 ======\n")
print(df[df['Overall Score'] < 80.0])

print("\n===== Release Date : Equal To : 2010 ======\n")
print(df[df['Release Date'] == 2010])

print("\n===== Data ======\n")

df
```
