### การสร้างตารางในฐานข้อมูล

```python
import pymysql

connection = pymysql.connect(
    host='localhost',
    user='u6090059',
    password='6090059',
    database='u6090059_mydb'
)
```

```python
cursor = connection.cursor()
cursor.execute('''
CREATE TABLE IF NOT EXISTS user (
    id INTEGER NOT NULL AUTO_INCREMENT,
    email VARCHAR(255) NOT NULL,
    fullname VARCHAR(255) NULL,
    PRIMARY KEY (id)
);
''')
cursor.execute('SHOW TABLES;')
results = cursor.fetchall()
for row in results:
    print(row[0])
```

### แก้ไขโครงสร้างของตาราง

```python
cursor = connection.cursor()
cursor.execute('''
ALTER TABLE user ADD city VARCHAR(255) NOT NULL AFTER fullname;
''')
cursor.execute('DESCRIBE user;')
results = cursor.fetchall()
for row in results:
    print(row)
```

### การเพิ่มข้อมูลในตาราง (INSERT)

```python
import pymysql
connection = pymysql.connect(
    host='localhost',
    user='u6090059',
    password='6090059',
    database='u6090059_mydb'
)
```

```python
cursor = connection.cursor()
cursor.execute('''
INSERT INTO user (email, fullname, city) VALUES (%s, %s, %s)
''', ('karn@me.com', 'Karn', 'Bangkok'))
connection.commit()
cursor.rowcount
```

```python
users = [
    ('ana@me.com', 'Ana', 'Phuket'),
    ('beer@me.com', 'Beer', 'Nonthaburi'),
    ('min@me.com', 'Min', 'Pathumthani')
]
cursor.executemany('''
INSERT INTO user (email, fullname, city) VALUES (%s, %s, %s)
''', users)
connection.commit()
cursor.rowcount
```

```python
import pandas as pd
df = pd.read_csv('user.csv')
users = [tuple(row) for row in df.itertuples(index=False)]
cursor.executemany('''
INSERT INTO user (email, fullname, city) VALUES (%s, %s, %s)
''', users)
connection.commit()
cursor.rowcount
```
