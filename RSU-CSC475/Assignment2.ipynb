```python
import matplotlib
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

plt.rcParams['font.family'] = "tahoma"

df = pd.read_csv('./assets/diag_code_2016.csv')
df['DATE_DIAG'] = pd.to_datetime(df['DATE_DIAG'])

df = df[df['DATE_DIAG'] >= '2016-07-01']
df = df[df['DATE_DIAG'] < '2016-10-01']
df = df.set_index('DATE_DIAG')

df[['A', 'B', 'C', 'D', 'N']].plot.area(alpha=0.4, figsize=(10,5))
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams['font.family'] = "tahoma"

diag = pd.read_csv("./assets/diag.csv")
person = pd.read_csv("./assets/person.csv")
province = pd.read_csv("./assets/province.csv")
    
diag = diag[["PID", 'CHRONIC']]
person = person[["PID", "PROVINCE_ID"]]

df = pd.merge(diag, person)
df = pd.merge(df, province)

df = df[df['CHRONIC'] == "E119"]

df["PROVINCE_NAME"].value_counts().plot.bar()
```
